```python
def task1(userInput):
    """
        –ï—Å–ª–∏ —Å–∫—Ä—ã—Ç–∞—è —Å—Ç—Ä–æ–∫–∞, –∫–æ—Ç–æ—Ä–∞—è –≤—Å—Ç—Ä–µ—á–∞–µ—Ç—Å—è —á–∞—â–µ –≤—Å–µ–≥–æ,
        –∏–º–µ–µ—Ç –¥–ª–∏–Ω—É –±–æ–ª—å—à–µ 2, —Ç–æ –¥–æ–ª–∂–Ω–∞ —Å—É—â–µ—Å—Ç–≤–æ–≤–∞—Ç—å —Å—Ç—Ä–æ–∫–∞, –∫–æ—Ç–æ—Ä–∞—è
        –≤—Å—Ç—Ä–µ—á–∞–µ—Ç—Å—è —Ä–æ–≤–Ω–æ —Å—Ç–æ–ª—å–∫–æ —Ä–∞–∑ –∏ –∏–º–µ–µ—Ç –¥–ª–∏–Ω—É —Ä–æ–≤–Ω–æ 2, —Ç.–∫.
        –º—ã –≤—Å–µ–≥–¥–∞ –º–æ–∂–µ–º –ø—Ä–æ—Å—Ç–æ –≤–∑—è—Ç—å –ø–µ—Ä–≤—ã–µ 2 –±—É–∫–≤—ã;
        
        –ü–æ—ç—Ç–æ–º—É –Ω–∞–º –Ω—É–∂–Ω–æ –ø—Ä–æ–≤–µ—Ä—è—Ç—å —Ç–æ–ª—å–∫–æ —Å—Ç—Ä–æ–∫–∏ –¥–ª–∏–Ω—ã 1 –∏ 2.
        –ü—Ä–æ–≤–µ—Ä—è–µ–º —Å—Ç—Ä–æ–∫–∏ –¥–ª–∏–Ω–æ–π 1 –ª–∏–Ω–µ–π–Ω–æ. –ß—Ç–æ–±—ã –ø—Ä–æ–≤–µ—Ä–∏—Ç—å
        —Å—Ç—Ä–æ–∫–∏ –¥–ª–∏–Ω—ã 2, –º—ã –º–æ–∂–µ–º –≤—ã–ø–æ–ª–Ω–∏—Ç—å –∏—Ç–µ—Ä–∞—Ü–∏—é –ø–æ S —Å–ª–µ–≤–∞ –Ω–∞–ø—Ä–∞–≤–æ –∏
        –¥–∏–Ω–∞–º–∏—á–µ—Å–∫–∏ –æ–±–Ω–æ–≤–ª—è—Ç—å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ä–∞–∑,
        –∫–æ–≥–¥–∞ –º—ã –≤–∏–¥–µ–ª–∏ –∫–∞–∂–¥—É—é —Å—Ç—Ä–æ–∫—É –¥–ª–∏–Ω—ã 1 –∏ 2.
    """

    arr = [0]*26
    arr1 = make_zeros(26, 26)

    for i in range(len(userInput)):
        current = ord(userInput[i]) - ord('a')


        for j in range(26):
            arr1[j][current] += arr[j]
        arr[current] += 1
    
    answer = 0;
    for i in range(26):
        answer = max(answer, arr[i])
    for i in range(26):
        for j in range(26):
            answer = max(answer, arr1[i][j])

    return answer
```

–ù–µ—Ç —Å–ø–æ—Å–æ–±–∞ –ø–æ–ª—É—á–∏—Ç—å ¬´bb¬ª –ø—Ä–∏ –ø–æ–ø—ã—Ç–∫–µ –Ω–∞–ø–µ—á–∞—Ç–∞—Ç—å ¬´ababa¬ª.

–ù–µ—Ç —Å–ø–æ—Å–æ–±–∞ –ø–æ–ª—É—á–∏—Ç—å ¬´aaaa¬ª –ø—Ä–∏ –ø–æ–ø—ã—Ç–∫–µ –Ω–∞–ø–µ—á–∞—Ç–∞—Ç—å ¬´aaa¬ª.

–ß—Ç–æ–±—ã –ø–æ–ª—É—á–∏—Ç—å ¬´ababa¬ª –ø—Ä–∏ –ø–æ–ø—ã—Ç–∫–µ –Ω–∞–ø–µ—á–∞—Ç–∞—Ç—å ¬´aababa¬ª, –º–æ–∂–Ω–æ –Ω–∞–∂–∞—Ç—å Backspace –≤–º–µ—Å—Ç–æ –ø–µ—á–∞—Ç–∏ –ø–µ—Ä–≤–æ–≥–æ —Å–∏–º–≤–æ–ª–∞, –∞ –∑–∞—Ç–µ–º –Ω–∞–ø–µ—á–∞—Ç–∞—Ç—å –≤—Å–µ –æ—Å—Ç–∞–≤—à–∏–µ—Å—è —Å–∏–º–≤–æ–ª—ã

```python
def task2(s, t):
    n = len(s)
    m = len(t)
    if m > n: return False

    p = (n - m)
    q = 0
    k = 0

    for i in range(p, n):
        if k == 1:
            k = 0
            continue
        if q < m and s[i] == t[q]:
            q += 1
        else:
            k += 1
        
    return q == m

print(task2('ababa', 'ba'), task2('ababa', 'bb'), task2('aaa', 'aaaa'), task2('aababa', 'ababa'))
```

```python
import sortedcontainers

def task5_1(array, minimize):
    left = sortedcontainers.SortedSet()
    for i in range(1, len(array)):
        left.add(array[i])
    for i in array:
        if i != -1:
            left.discard(i)
    lastPlaced = -1
    for i in range(len(array)):
        if array[i] == -1:
            if minimize:
                it = left.pop(0)
            else:
                it = left.pop(left.bisect_left(lastPlaced) - 1)
            array[i] = it
        else:
            lastPlaced = array[i]

        


def task5():
    n = int(input('–î–ª–∏–Ω–∞?: '))
    q = []
    string = input('–í—Ö–æ–¥–Ω–æ–π –º–∞—Å—Å–∏–≤: ')
    for i in string.split(' '):
        q.append(int(i))

    res1 = [-1] * n
    res2 = [-1] * n

    lastPlaced = -1
    for i in range(n):
        if lastPlaced != q[i]:
            res1[i] = q[i]
            res2[i] = q[i]
        lastPlaced = q[i]
    task5_1(res1, True)
    task5_1(res2, False)

    print(res1, '\n', res2)

def main():
    t = int(input('–°–∫–æ–ª—å–∫–æ —Ä–∞–∑ —Ç–µ—Å—Ç–∏–º?: '))
    while t > 0:
        task5()
        t -= 1

main()
```
